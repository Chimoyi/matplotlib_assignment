Load the dataset into a Pandas DataFrame and explore the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = '''Month,Revenue,Orders,Average_Order_Value
Jan-2020,10000,200,50
Feb-2020,11000,220,50
Mar-2020,12000,240,50
Apr-2020,13000,260,50
May-2020,14000,280,50
Jun-2020,15000,300,50
Jul-2020,16000,320,50
Aug-2020,17000,340,50
Sep-2020,18000,360,50
Oct-2020,19000,380,50
Nov-2020,20000,400,50
Dec-2020,21000,420,50
Jan-2021,22000,440,50
Feb-2021,23000,460,50
Mar-2021,24000,480,50
Apr-2021,25000,500,50
May-2021,26000,520,50
Jun-2021,27000,540,50
Jul-2021,28000,560,50
Aug-2021,29000,580,50
Sep-2021,30000,600,50
Oct-2021,31000,620,50
Nov-2021,32000,640,50
Dec-2021,33000,660,50'''

data1 = "jumta.csv"
df = pd.read_csv(data1)
```

Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality.

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Revenue'], marker='o', linestyle='--', linewidth=2, label='Revenue', color = 'green')
plt.title('Monthly Revenue (2020-2021)')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

Create a bar plot showing the total number of orders per month. Customize the plot to make it publication-quality.

```python
plt.bar(df['Month'], df['Orders'], color = 'purple', label = 'Orders')
plt.title('Orders per Month (2020-2021)')
plt.xlabel('Month')
plt.ylabel('No. of Orders')
plt.xticks(rotation=45)
plt.grid(axis = 'y')
plt.legend()
plt.show()
```

Create a scatter plot showing the relationship between the number of orders and the average order value. Customize the plot to make it publication-quality.



```python
plt.scatter(df['Orders'], df['Average_Order_Value'], color = 'blue', label = 'Orders')
plt.title('Number of Orders vs Average Order Value (2020-2021)')
plt.xlabel('No. of Orders')
plt.ylabel('Av. Order Value')
plt.xticks(rotation=45)
plt.grid(axis = 'y')
plt.legend()
plt.show()
```

Create a subplot containing all three visualizations created in steps 2-4. Adjust the layout and size of the subplot to make it visually appealing.

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

ax1.plot(df['Month'], df['Revenue'], marker='o', linestyle='-', linewidth=2, label='Revenue')
ax1.set_title('Monthly Revenue (2020-2021)')
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue')
ax1.tick_params(axis='x', rotation=45)
ax1.grid()
ax1.legend()

ax2.bar(df['Month'], df['Orders'], color='orange', label='Orders')
ax2.set_title('Monthly Orders (2020-2021)')
ax2.set_xlabel('Month')
ax2.set_ylabel('Number of Orders')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(axis='y')
ax2.legend()

ax3.scatter(df['Orders'], df['Average_Order_Value'], color='green', label='Order Value')
ax3.set_title('Number of Orders vs Average Order Value (2020-2021)')
ax3.set_xlabel('Number of Orders')
ax3.set_ylabel('Average Order Value')
ax3.grid()
ax3.legend()
```
